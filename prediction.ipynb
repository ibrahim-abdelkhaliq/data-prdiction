```python
import csv
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df.year,df.per_capita_income,color='blue',marker='+')
```

```python
year = df.drop('per_capita_income',axis='columns')
year
```

```python
per_capita_income = df.per_capita_income
per_capita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
